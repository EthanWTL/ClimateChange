```python
X_train = np.concatenate((X_CL, X_COH, X_COL, X_NROI),axis = 0)
X_train.shape
```

```python
Y_train = np.concatenate((Y_CL, Y_COH, Y_COL, Y_NROI),axis = 0)
Y_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers import Dropout
from keras.layers.convolutional import Conv1D
from keras.layers.convolutional import MaxPooling1D
from tensorflow.keras.utils import to_categorical
```

```python
model = Sequential()
model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(X_train.shape[1:])))
model.add(Dropout(0.5))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(32, activation='relu'))
model.add(Dense(4, activation='softmax'))
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
import time
start = time.time()
print("Timer Begin")
model.fit(X_train,Y_train,batch_size=4,epochs=100)

end = time.time()
print(end - start)
```

```python
Y_pred1 = model.predict(X_train)
Y_pred1=Y_pred1.argmax(axis=1)
```

```python
from sklearn.metrics import confusion_matrix
rf_conf_mat = confusion_matrix(Y_train, Y_pred1)
print("Confusion matrix: \n", rf_conf_mat)
```

```python
from sklearn.metrics import classification_report
CR1 = classification_report(Y_train,Y_pred1, digits= 4)
print(CR1)
```
