```python
print(x.max())
print(x.min())
print(y.max())
print(y.min())
```

```python
normalized_x = (x-x.min())/(x.max()-x.min())
normalized_y = (y-y.min())/(y.max()-y.min())
```

```python
first = np.where(normalized_x >= 0.5,1,0)
```

```python
df = pd.DataFrame(first,columns=['first'])
```

```python
def sec(num):
    if num >= 0.25 and num < 0.5:
        return 1
    if num >= 0.75 and num <= 1:
        return 1
    else:
        return 0
```

```python
df['second'] = normalized_x.apply(lambda x: sec(x))
```

```python
thr = np.where(normalized_y>=0.5,1,0)
df['thrid']=thr
```

```python
df['Forth'] = normalized_y.apply(lambda x: sec(x))
```
