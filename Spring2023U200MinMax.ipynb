```python
from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing
import numpy as np
import csv
```

### MinMax

```python
path = "reverse_csv"
print("Files and directories in a specified path:")
for filename in os.listdir(path):
    f = os.path.join(path,filename)
    if os.path.isfile(f): 
        f_name = 'reverse_csv/' + f[12:]
        f_name_new = f_name[:-4]
        #print(f_name_new)
    record = pd.read_csv(f_name, header = None)
    data = record.values
    
    #print(data)
    # define min max scaler
    scaler = MinMaxScaler()
    # transform data
    scaled = scaler.fit_transform(data)
    #print(scaled)
    
    
    np.set_printoptions(suppress=True)
    np.set_printoptions(precision=6)
    np.savetxt(fname = "WholeDataCSV/" + f_name_new[12:] + ".csv", delimiter=",", fmt = '%.06f',X=scaled)
```

### Standard

```python
from sklearn.preprocessing import StandardScaler
```

```python
path = "reverse_csv"
print("Files and directories in a specified path:")
for filename in os.listdir(path):
    f = os.path.join(path,filename)
    if os.path.isfile(f): 
        f_name = 'reverse_csv/' + f[12:]
        f_name_new = f_name[:-4]
        #print(f_name_new)
    record = pd.read_csv(f_name, header = None)
    data = record.values
    
    #print(data)
    # define standard scaler
    scaler = StandardScaler()
    # transform data
    scaled = scaler.fit_transform(data)
    #print(scaled)
    
    
    np.set_printoptions(suppress=True)
    np.set_printoptions(precision=6)
    np.savetxt(fname = "WholeDataCSV_Standard/" + f_name_new[12:] + ".csv", delimiter=",", fmt = '%.06f',X=scaled)
```

```python
data1 = pd.read_table("InterestImageCenter.txt",sep = '_', header = None)
data1.columns = ['Day', 'Cluster','ConectedComponent','CenterX','CenterY']
data1
```

### RawInterestData15*15

```python
a = data1.shape[0]
for i in range(a):
    day = data1['Day'][i]
    c = data1['Cluster'][i]
    cp = data1['ConectedComponent'][i]
    #print(day)
    df = pd.read_csv('reverse_csv/day'+ str(day) +'.csv', header = None)
    df1 = df.values
    cx = data1['CenterX'][i]
    cy = data1['CenterY'][i]
    #print(cx)
    #print(cy)
    #left
    x_left = cx-8
    #right
    x_right = cx+7  
    #top
    y_top = cy-8  
    #bottom
    y_bottom = cy+7
    df2 = df1[y_top:y_bottom,x_left:x_right]
    #df2
    df3 = pd.DataFrame(df2)
    #df3
    df3.to_csv('./RawInterestImage_CSV/' +  'Day' + str(day) + '_' + str(c) + '_' + str(cp) +  '_'+ str(cx) + '_' + str(cy) + '.csv', header = 0, encoding='utf-8', index=False)
```
